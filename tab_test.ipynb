```python
%pip install pdfplumber
import pdfplumber
import pandas as pd

# Replace 'your_pdf_file.pdf' with the actual path to your PDF file
pdf_path = 'AIR-Dienstplan_2024_03.pdf'

# Open the PDF file
with pdfplumber.open(pdf_path) as pdf:
    # Extract tables from all pages
    all_tables = []
    for page in pdf.pages:
        # Extract table data as a list of lists
        table_data = page.extract_tables()
        all_tables.extend(table_data)

# Convert the table data to DataFrames without an index
dfs = [pd.DataFrame(table) for table in all_tables]
```

```python
# Display the entire DataFrame without truncation
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

df = dfs[0]

column_name = 0

# Filter rows where the value in the first column is not None or the first occurrence
filtered_df = df[~((df[column_name].isnull()) & (df.duplicated(subset=[column_name])))]
filtered_df.reset_index(drop=True, inplace=True)
df2 = filtered_df

# rename columns
df2.columns = df2.iloc[0]
df2 = df2.iloc[1:].reset_index(drop=True)

# Extract the first column into a separate DataFrame
df3 = df2.iloc[:, [0]].copy()

# Drop columns with the name 'None'
df2 = df2.drop(columns=[col for col in df2.columns if col == None])

# Concatenate along the columns (axis=1)
df4 = pd.concat([df3, df2], axis=1)

# Rename the first column
df4 = df4.rename(columns={df4.columns[0]: "name"})

df
```
